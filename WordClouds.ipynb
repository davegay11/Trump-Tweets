# Trump Wordcloud

> We create a wordcloud of Trump's most common words

```python
from wordcloud import WordCloud
import os
from PIL import Image
import pandas as pd
import numpy as np

# Show matplotlib inline
%matplotlib inline
```

```python
d = os.path.dirname(os.path.realpath('__file__'))

# Here we read in the cleaned text and then concatenate everything together
# so that wordcloud can read it
df = pd.read_csv("./clean_data/clean_tweets.csv", header=0)
```

```python
text = df['clean_text'].str.cat(sep=' ')
```

```python
# Let's first just make a simple wordcloud showing the most frequent
# words that show up in Donald Trump's Tweets
wordcloud = WordCloud().generate(text)
wordcloud.to_file(path.join(d, 'img/basic_wordcloud.png'))

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# Now let's have some fun...

# read the mask image
trump_mask = np.array(Image.open(path.join(d, "img/trump_outline.png")))
wc = WordCloud(background_color="white", max_words=2000, mask=trump_mask)
# generate word cloud
wc.generate(text)

# store to file
wc.to_file(path.join(d, "img/trump_wordcloud.png"))

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(trump_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
